```python
from data_utils.stochastic_utils.basic_distributions import NormalDistribution,LogNormalDistribution,WeibullDistribution
from data_utils.serial_utils.series_trans_utils import MinMax,ZScore
from data_utils.serial_utils.data_series import DataSeries
import numpy
from matplotlib import pyplot
```

```python
wd = WeibullDistribution(2, 5)
```

```python
rwd = wd.rvf(100)
sorted_index = numpy.argsort(rwd)
nd = NormalDistribution(numpy.mean(rwd), numpy.std(rwd, ddof=1))
q = nd.ppf(first=0.01, end=0.99, num=100).y
```

```python
pyplot.plot(rwd[sorted_index])
pyplot.plot(q)
pyplot.show()
```

```python
diff = rwd[sorted_index] - q
pyplot.plot(diff)
pyplot.show()
```

```python
nwd = [0] * len(rwd)
diff_ = []
diff_index = []
# for i,r in enumerate(rwd[sorted_index]):
for i in range(len(rwd)):
    nwd[sorted_index[i]] = q[i]
nwd = numpy.array(nwd)
```

```python
pyplot.plot(numpy.sort(rwd))
pyplot.plot(numpy.sort(nwd))
pyplot.plot(q)
pyplot.legend(["rwd", "nwd", "q"])
pyplot.show()

wd_diff = rwd-nwd
pyplot.plot(wd_diff[sorted_index])
pyplot.plot(diff)
pyplot.plot(diff_)
pyplot.show()
```
